```python
import processor
```

```python
Processor = processor.AscProcessor()
```

```python
train = {}
val = {}
test = {}
for dataset in DATASETS:
    domain_name = dataset.split('/')[-1]
    train[domain_name] = Processor.get_train_examples(dataset)
    val[domain_name] = Processor.get_dev_examples(dataset)
    test[domain_name] = Processor.get_test_examples(dataset)
```

```python
print('train:', train)
```

```python
import pickle
with open('./../data/train.pkl', 'wb') as f:
    pickle.dump(train, f)
with open('./../data/val.pkl', 'wb') as f:
    pickle.dump(val, f)
with open('./../data/test.pkl', 'wb') as f:
    pickle.dump(test, f)
```
